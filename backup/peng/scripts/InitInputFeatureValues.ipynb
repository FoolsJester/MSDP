```python
# This script will generate input feature values on particular line (such as 39A). 
# For each line, there are 2 direictions, dirction 1 and direction 2

import csv
import pandas as pd
import numpy as np
sample_df = pd.read_csv("LeaveTimesTrips_JUN.csv")
sample_df.count()
#sample_df.head()
```

```python
# 39A, direction 1 or 2
sample_df = sample_df[(sample_df['LineId']=='39A') & (sample_df['Direction'] == 2)]
sample_df.head()
```

```python
tripids = sample_df.TripId.sort_values().unique()
print(tripids)
```

```python
PN = sample_df.ProgrNumber.sort_values().unique()
```

```python
len(sample_df[(sample_df['TripId'] == 4117250) & (sample_df['ProgrNumber'] == 2)])
```

```python
useful_tripids = []
for tripid in tripids:
    isUsefull = True
    for pnumber in PN:
        if len(sample_df[(sample_df['TripId'] == tripid) & (sample_df['ProgrNumber'] == pnumber)]) == 0:
            isUsefull = False
            break
    if isUsefull:
        useful_tripids.append(tripid)
```

```python
stopPoints = []
for i in PN:
    stopid = sample_df[(sample_df['TripId'] == 5002342) & (sample_df['ProgrNumber'] == i)].StopPointId.iloc[0]
    stopPoints.append(stopid)
```

```python
count = 1
stopTable = {}
for stopid in stopPoints:
    stopTable['{}_P'.format(str(count).zfill(3))] = stopid
    count += 1
```

```python
df = pd.DataFrame(stopTable, index=[0])
df['tripid'] = '0'
```
